```python
def cadastro():
    hospedes = {}  # Dicion√°rio para armazenar nome e di√°rias

    for i in range(5):
        nome = input(f"Digite o nome do {i+1}¬∫ h√≥spede: ")

        # Verifica se o nome j√° existe
        while nome in hospedes:
            print("Esse nome j√° foi cadastrado. Digite um nome diferente.")
            nome = input(f"Digite o nome do {i+1}¬∫ h√≥spede: ")

        diarias = int(input(f"Digite a quantidade de di√°rias de {nome}: "))

        # Armazena no dicion√°rio
        hospedes[nome] = diarias

    print("\nCadastro conclu√≠do!")
    print("Lista de h√≥spedes e suas di√°rias:")
    for nome, diarias in hospedes.items():
        print(f"{nome}: {diarias} di√°rias")

# Chamada da fun√ß√£o
cadastro()
```

```python
def calcular_custo_hospedagem():
    try:
        nome_hospede = input("Digite o nome do h√≥spede: ")
        quantidade_diarias_str = input("Digite a quantidade de di√°rias: ")

        # Tratamento de erro para entrada n√£o num√©rica
        quantidade_diarias = int(quantidade_diarias_str)

        if quantidade_diarias < 1:
            print("A quantidade de di√°rias deve ser maior que zero.")
            return

        taxa_adicional = 0
        if quantidade_diarias < 15:
            taxa_adicional = 20.00
        elif quantidade_diarias == 15:
            taxa_adicional = 14.00
        else: # quantidade_diarias > 15
            taxa_adicional = 12.00

        custo_total = (300 * quantidade_diarias) + (taxa_adicional * quantidade_diarias)

        print(f"\nNome do h√≥spede: {nome_hospede}")
        print(f"Valor total a pagar: R$ {custo_total:.2f}")

    except ValueError:
        print("Erro: A quantidade de di√°rias deve ser um n√∫mero inteiro.")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

# Chama a fun√ß√£o para executar o c√°lculo
calcular_custo_hospedagem()
```

```python
biblioteca = {}

def adicionar_livro():
    try:
        t = input("T√≠tulo: ").title()
        if t in biblioteca: raise ValueError("Livro j√° cadastrado.")
        a = input("Autor: ").title()
        biblioteca[t] = {"autor": a, "emprestado": False}
        print("‚úÖ Adicionado!")
    except Exception as e:
        print("Erro:", e)

def emprestar_livro():
    try:
        t = input("T√≠tulo: ").title()
        if t not in biblioteca: raise KeyError("N√£o encontrado.")
        if biblioteca[t]["emprestado"]: raise Exception("J√° emprestado.")
        biblioteca[t]["emprestado"] = True
        print("üìö Emprestado!")
    except Exception as e:
        print("Erro:", e)

def devolver_livro():
    try:
        t = input("T√≠tulo: ").title()
        if t not in biblioteca: raise KeyError("N√£o encontrado.")
        if not biblioteca[t]["emprestado"]: raise Exception("N√£o estava emprestado.")
        biblioteca[t]["emprestado"] = False
        print("üìñ Devolvido!")
    except Exception as e:
        print("Erro:", e)

def listar_livros_disponiveis():
    disp = [t for t, i in biblioteca.items() if not i["emprestado"]]
    print("\nLivros dispon√≠veis:" if disp else "\nNenhum dispon√≠vel.")
    for t in disp:
        print(f"- {t} ({biblioteca[t]['autor']})")

def buscar_livro():
    try:
        t = input("T√≠tulo: ").title()
        if t not in biblioteca: raise KeyError("N√£o encontrado.")
        i = biblioteca[t]
        s = "Emprestado" if i["emprestado"] else "Dispon√≠vel"
        print(f"{t} - {i['autor']} ({s})")
    except Exception as e:
        print("Erro:", e)

while True:
    print("\n1.Adicionar  2.Emprestar  3.Devolver  4.Listar  5.Buscar  0.Sair")
    try:
        op = int(input("Op√ß√£o: "))
        if op == 1: adicionar_livro()
        elif op == 2: emprestar_livro()
        elif op == 3: devolver_livro()
        elif op == 4: listar_livros_disponiveis()
        elif op == 5: buscar_livro()
        elif op == 0: break
        else: print("Inv√°lido.")
    except ValueError:
        print("Digite um n√∫mero.")
```

```python
oficina = {"veiculos": {}, "servicos": []}

def reg_veiculo():
    p = input("Placa: ").upper()
    if p in oficina["veiculos"]:
        print("Ve√≠culo j√° cadastrado.")
        return
    dono = input("Dono: ")
    modelo = input("Modelo: ")
    oficina["veiculos"][p] = {"dono": dono, "modelo": modelo}
    print("Ve√≠culo adicionado!")

def reg_servico():
    p = input("Placa: ").upper()
    if p not in oficina["veiculos"]:
        print("Ve√≠culo n√£o encontrado.")
        return
    s = input("Servi√ßo: ")
    oficina["servicos"].append((p, s))
    print("Servi√ßo registrado!")

def listar_veiculos():
    if not oficina["veiculos"]:
        print("Nenhum ve√≠culo cadastrado.")
    else:
        for p, v in oficina["veiculos"].items():
            print(f"{p} - {v['modelo']} ({v['dono']})")

def listar_servicos():
    if not oficina["servicos"]:
        print("Nenhum servi√ßo registrado.")
    else:
        for p, s in oficina["servicos"]:
            print(f"{p} - {s}")

def buscar():
    p = input("Placa: ").upper()
    if p in oficina["veiculos"]:
        v = oficina["veiculos"][p]
        print(f"{p} - {v['modelo']} ({v['dono']})")
    else:
        print("Ve√≠culo n√£o encontrado.")

while True:
    print("\n1.Reg. ve√≠culo 2.Reg. servi√ßo 3.Listar ve√≠culos 4.Listar servi√ßos 5.Buscar 0.Sair")
    o = input("Op√ß√£o: ")
    if o == "1": reg_veiculo()
    elif o == "2": reg_servico()
    elif o == "3": listar_veiculos()
    elif o == "4": listar_servicos()
    elif o == "5": buscar()
    elif o == "0": break
    else: print("Op√ß√£o inv√°lida.")
```
